# SVD of Rank Deficient Matrix

```python
import numpy as np
```

```python
M = np.array([[1,0,3],
    [3,7,2],
    [2,-2,8],
    [0,-1,1],
    [5,8,7],
    ])
M
```

```python
A = M.T @ M
A
```

```python
B = M @ M.T
B
```

## Eigenvalues

```python
eigvals, eigvecs = np.linalg.eig(A)
eigvals, eigvecs
```

```python
eigvals, eigvecs = np.linalg.eig(B)
eigvals, eigvecs
```

## SVD

```python
eigvals, eigvecs = np.linalg.eig(A)
V = eigvecs[:, (0,2)]
eigvals, eigvecs = np.linalg.eig(B)
U = eigvecs[:, (0,2)]
M = U @ np.array([[-np.sqrt(2.14670489e+02), 0],[0,-np.sqrt(6.93295108e+01)]]) @ V.T
M
```

## 1-D Approximation

```python
V = V[:,0:1]
U = U[:,0:1]
N = U @ np.array([[-np.sqrt(2.14670489e+02)]]) @ V.T
N
```

```python
eigvals, eigvecs = np.linalg.eig(A)
V = eigvecs[:, (0,2)]
eigvals, eigvecs = np.linalg.eig(B)
U = eigvecs[:, (0,2)]
M = U @ np.array([[-np.sqrt(2.14670489e+02), 0],[0,0]]) @ V.T
M
```
